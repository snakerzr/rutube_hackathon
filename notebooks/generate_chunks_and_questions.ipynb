```python
client = OpenAI()
def complete(query: str, instruction: str):
    completion = client.chat.completions.create(
        model="gpt-4o-mini",
        temperature=1,
        messages=[
            {"role": "system", "content": instruction},
            {
                "role": "user",
                "content": query
            }
        ]
    )
    return completion

def extract_content(completion) -> str:
    try:
        return completion.choices[0].message.content
    except:
        return '==ERROR=='
```

```python
SYSTEM_PROMPT = '''Ты клиент сервиса RUTUBE, обращаешься в службу поддержки.
Напиши вопрос, на который можно ответить по этому куску. 
Если по этому куску сложно сгенерировать такой вопрос, который бы задавал клиент в службу поддержки, то напиши ПРОПУСКАЮ.
Пиши только вопрос или ПРОПУСКАЮ.
'''.strip()

INSTRUCTION = '''
КУСОК:
=====
{context}
=====
'''.strip()

result = []
for text in texts:
    answer = extract_content(complete(INSTRUCTION.format(context=text.page_content),SYSTEM_PROMPT))
    result.append([text,answer])
```
